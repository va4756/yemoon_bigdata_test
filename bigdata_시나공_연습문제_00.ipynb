### 12.결측치 처리, 최빈값, 데이터 개수(p161)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 결측치가 가장 많은 컬럼, 출력 결과로 확인
print(df.isnull().sum())
df['f3'].value_counts()
```

```python
# 결측치가 가장 많은 컬럼 제거
df = df.dropna(subset=['f1'])

# 두번째로 가장 결측치가 많은 컬럼 최빈값 대체
freq = df['f3'].mode()[0]
df['f3'].fillna(freq, inplace=True)

# f3컬럼이 gold인 데이터의 수
print(sum(df['f3'] == 'gold'))
```

```python
# 방법2
# 결측치가 가장 많은 컬럼
col = df.isnull().sum().idxmax()

# 결측치가 가장 많은 컬럼 제거
df = df.dropna(subset=[col])

# 두번째로 가장 결측치가 많은 컬럼 최빈값 대체
freq = df['f3'].mode()[0]
df['f3'] = df['f3'].fillna(freq)

# f3컬럼이 gold인 데이터의 수
print(len(df[df['f3'] == 'gold']))
```

### 16.이상치, IQR(p167)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 3사분위 1사분위
q1 = df['views'].quantile(0.25)
q3 = df['views'].quantile(0.75)

# IQR 계산
IQR = q3 - q1

# 이상치 기준 찾기
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

# 이상치 조건에 만족하는 데이터 찾기
cond1 = df['views'] < lower_bound
cond2 = df['views'] > upper_bound
df = df[cond1 | cond2]

# 이상치 데이터의 views 컬럼 합
print(int(df['views'].sum()))
```

### 17.이상치, 소수점 있는 데이터 찾기, 표준편차(p168)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 표준편차
view_std_1 = df['views'].std()
print(view_std_1)

# 음수이거나 0인 테이터 제외(age)
cond = df['age'] <= 0
df = df[~cond]

# 소수점이 없는 값만 선택
cond = df['age'] == round(df['age'], 0)
df = df[cond]

# 표준편차
view_std_2= df['views'].std()

print(round((view_std_1 + view_std_2), 2))
```

### 18.데이터(행) 기준 평균값, 인덱싱(p169)

```python
import pandas as pd
# df = pd.read_csv("type1_data2.csv", index_col="year")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data2.csv", index_col="year")
df.head()
```

```python
# 2001에서, 평균보다 큰값 합계
# (참고) sum()은 판다스에서는 기본적으로 컬럼별 계산을 한다.
m1 = df.loc[2001].mean()
cond1 = df.loc[2001] > m1
r1 = sum(cond1)
print(r1)

# 2003에서, 평균보다 작은값 합계
m2 = df.loc[2003].mean()
cond2 = df.loc[2003] < m2
r2 = sum(cond2)
print(r2)
print('*' * 50)
print(r1 + r2)
```

```python
# loc를 사용하기 싫으면 T를 사용할 것
df = df.T

m1 = df[2001].mean()
cond1 = df[2001] > m1
r1 = sum(cond1)

m2 = df[2003].mean()
cond2 = df[2003] < m2
r2 = sum(cond2)

print(r1 + r2)
```

### 19.결측치(뒤의 값으로 대체), 그룹합(p171)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 결측치 채우기
df = df.fillna(method='bfill')

# city와 f2기준 그룹합 계산
df = df.groupby(['city', 'f2']).sum(numeric_only=True).reset_index()
print(df.head())

# 내림차순 정렬
df = df.sort_values('views', ascending=False)

# 3번째로 큰 값 출력
print(df.iloc[2, 0])
```

### 20.시계열 데이터, 월별 집계, 인덱스(p172)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```
