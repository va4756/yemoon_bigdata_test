### 04.값 변경, 정렬, 합계(p151)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 결측 데이터 0으로 대체
df['views'].fillna(0, inplace=True)   # 할당하지 말 것
print(df.head())

# 내림차순 정렬
df = df.sort_values(by='views', ascending=False)

# views 상위 10개 데이터에 최소값 대입
views_min = df.iloc[:10]['views'].min()
df.loc[:10, 'views'] = views_min

# views 컬럼의 합
result = int(df['views'].sum())
print('*'*50)
print(result)
```

### 06.필터링, 분산(p154)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 조건에 맞는 데이터
cond1 = df['f3'] == 'gold'
cond2 = df['f2'] == 2
df = df[cond1 & cond2]

# 분산
result = df['f1'].var()

# 출력
print(round(result, 2))
```

### 07.값 변경(연산), 필터링 절댓값

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 나이 + 1
df['age'] = df['age'] + 1

# 20대, 30대 조건
cond1 = (df['age'] >= 20) & (df['age'] < 30)
cond2 = (df['age'] >= 30) & (df['age'] < 40)

# 20대와 30대 views 평균의 차이(절대값)
diff = abs(df[cond1]['views'].mean() - df[cond2]['views'].mean())
print(round(diff, 2))
```

### 08.시계열 데이터, 필터링, 데이터 개수

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 자료형 변환
df['subscribed'] = pd.to_datetime(df['subscribed'])

# 파생변수 생성 (년, 월)
df['year'] = df['subscribed'].dt.year
df['month'] = df['subscribed'].dt.month

# 2024년 2월이고, 'f3'이 gold인 데이터
cond1 = df['year'] == 2024
cond2 = df['month'] == 2
cond3 = df['f3'] == 'gold'
df = df[cond1 & cond2 & cond3]

# 데이터 개수
print(df)
print('*'*50)
print(len(df))
```
