## 제 1유형 문제

### 문1-1) Boston Housing 데이터셋에서
-  상위 10번째 값으로 상위 10개의 값을 변환하고
- age가 80이상인 데이터를 추출하고
- crim의 평균값 계산

```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv("/content/yemoonsaBigdata/datasets/Part5/201_boston.csv")
df.head()
```

```python
# 내림차순으로 정렬(CRIM)
df = df.sort_values(by='CRIM', ascending=False)
df.head(10)
```

```python
# 특정 위치 데이터 찾기(CRIM의 10번째 값)
crim_val = df.iloc[9, 0]  # [10번째 행, 1번째 열] = [9, 0]
crim_val
```

```python
# 특정 위치 데이터 변경하기
df.iloc[:10, 0] = crim_val
df.head(11)
```

```python
# 조건부 평균 계산하기  # 소수 셋째자리까지
cond = df[df['AGE'] >= 80]
result = cond['CRIM'].mean()
print(round(result, 3))
```

### 문1-2) Californai Housing 데이터셋에서
- 순서대로 80%의 데이터를 추출하고
- 'total_bedrooms' 변수의 중앙값을 구하고
- 'total_bedrooms'의 결측치에 이 값으로 대체한 후,
- 결측치 대체 전과 후의 표준편차를 구하고
- 이 둘의 차를 절대값으로 구하시오.

```python
import pandas as pd

df = pd.read_csv("/content/yemoonsaBigdata/datasets/Part5/202_housing.csv")
df.head()
```

```python
df.info()
```

```python
# 데이터 추출하기
val_80 = int(len(df) * 0.8)
rst_df = df.iloc[:val_80]['total_bedrooms']
rst_df
```

```python
# 표준편차, 중앙값 계산하기
rst_std = rst_df.std()
rst_med = rst_df.median()
rst_std, rst_med
```

```python
# 결측값 채우기 & 표준편차 계산하기
rst_df_fillna = rst_df.fillna(rst_med)
rst_std_2 = rst_df_fillna.std()
rst_std_2
```

```python
# 결과 출력하기
result = abs(rst_std - rst_std_2)
result
```

### 문1-3) (문1-2)의 Californai Housing 데이터셋에서
- 'population' 컬럼에서 이상치의 갯수를 구하시오
- 이상치를 구하기 위해서는 평균, 1.5X(표준편차) 를 이용한다.

```python
import pandas as pd

df = pd.read_csv("/content/yemoonsaBigdata/datasets/Part5/202_housing.csv")
df.head()
```

```python
mean = df['population'].mean()
std = df['population'].std()
print(mean, std)
```

```python
lower_bound = mean - 1.5 * std
upper_bound = mean + 1.5 * std

res_df = df.loc[(df['population'] < lower_bound) | (df['population'] > upper_bound), 'population']
result = res_df.sum()
result
```

## 제 2유형 문제

### (문제) E-Commerce Shipping 데이터셋이다. 주어진 훈련데이터셋을 이용하여 고객이 주문한 물품의 정시 도착 여부를 예측하고 예측 결과를 CSV 파일로 제출하시오.
- 예측값의 ROC-AUC 점수 결과로 점수를 매김

```python
import pandas as pd
X_test = pd.read_csv("yemoonsaBigdata/datasets/Part5/204_x_test.csv")
X_train = pd.read_csv("yemoonsaBigdata/datasets/Part5/204_x_train.csv")
y_train = pd.read_csv("yemoonsaBigdata/datasets/Part5/204_y_train.csv")
```

#### 2.각각의 열에 대해 분리작업

```python
COL_DEL = ["ID"]
COL_NUM = ['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']
COL_CAT = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']
COL_Y = ['Reached.on.Time_Y.N']
```

#### 3.데이터 모형 구축

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train[COL_NUM+COL_CAT],
                                                  y_train[COL_Y],
                                                  test_size=0.3,
                                                  stratify=y_train[COL_Y])
X_tr.shape, X_val.shape, y_tr.shape, y_val.shape
```

```python
y_train
```

#### 4.스케일링 작업 시행

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_tr[COL_NUM])

X_tr[COL_NUM] = scaler.transform(X_tr[COL_NUM])
X_val[COL_NUM] = scaler.transform(X_val[COL_NUM])
X_test[COL_NUM] = scaler.transform(X_test[COL_NUM])
```

```python
X_tr[COL_NUM].head()
```

#### 5.인코딩 작업 실시
- 범주형 변수에 대해

```python
from sklearn.preprocessing import LabelEncoder

X = pd.concat([X_train[COL_CAT], X_test[COL_CAT]])

for col in COL_CAT:
    le = LabelEncoder()
    le.fit(X[col])
    X_tr[col] = le.transform(X_tr[col])
    X_val[col] = le.transform(X_val[col])
    X_test[col] = le.transform(X_test[col])

    print(col)
    print(le.classes_)
    print('\n')
```

```python
X_tr[COL_CAT].head()
```

#### 6.모형학습

```python
# 1) 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(random_state=123)
model_rf.fit(X_tr, y_tr)

# 2) XGBoost
from xgboost import XGBClassifier

model_xgb = XGBClassifier()
model_xgb.fit(X_tr, y_tr)
```

#### 7. 데이터 모형 평가

##### 7.1 검증 데이터로 예측값 생성

```python
X_val.head()
```

```python
y_pred_rf = model_rf.predict(X_val)[:, 1]
y_pred_xgb = model_xgb.predict(X_val)[:, 1]
```

##### 7.2 평가지표 구하기

```python
## 답안 채점 기준인 roc_auc 사용
from sklearn.metrics import roc_auc_score

score_rf = roc_auc_score(y_val, y_pred_rf)
score_xgb = roc_auc_score(y_val, y_pred_xgb)

print('Random Forest: ', score_rf)
print('XGBoost: ', score_xgb)
```

##### 7.3 학습, 검증 성능 확인 함수

```python
def get_scores(model, X_tr, X_val, y_tr, y_val):

  y_tr_pred = model.predict_proba(X_tr)[:, 1]
  y_val_pred = model.predict_proba(X_val)[:, 1]
  tr_score = roc_auc_score(y_tr, y_tr_pred)
  val_score = roc_auc_score(y_val, y_val_pred)

  return f'train: {round(tr_score, 4)}, valid: {round(val_score, 4)}'
```

```python
print('Random Forest \t-', get_scores(model_rf, X_tr, X_val, y_tr, y_val))
print('XGBoost \t-', get_scores(model_xgb, X_tr, X_val, y_tr, y_val))
```

##### 7.4 하이퍼 파라미터 튜닝

```python
modelXGB2 = XGBClassifier(n_estimators=30, max_depth=3, min_child_weight=1, random_state=123)
modelXGB2.fit(X_tr, y_tr)

modelXGB3 = XGBClassifier(n_estimators=30, max_depth=5, min_child_weight=1, random_state=123)
modelXGB3.fit(X_tr, y_tr)

modelXGB4 = XGBClassifier(n_estimators=50, max_depth=3, min_child_weight=1, random_state=123)
modelXGB4.fit(X_tr, y_tr)

modelXGB5 = XGBClassifier(n_estimators=50, max_depth=5, min_child_weight=1, random_state=123)
modelXGB5.fit(X_tr, y_tr)

print('XGBoost2 \t-', get_scores(modelXGB2, X_tr, X_val, y_tr, y_val))
print('XGBoost3 \t-', get_scores(modelXGB3, X_tr, X_val, y_tr, y_val))
print('XGBoost4 \t-', get_scores(modelXGB4, X_tr, X_val, y_tr, y_val))
print('XGBoost5 \t-', get_scores(modelXGB5, X_tr, X_val, y_tr, y_val))
```

##### 7.5 최종 결과값 생성

```python
pred = modelXGB5.predict_proba(X_test[COL_NUM+COL_CAT])[:, 1]
result = pd.DataFrame({'ID': X_test['ID'], 'pred': pred})
print(result.head())
```
