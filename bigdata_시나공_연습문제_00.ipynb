### 16.이상치, IQR(p167)

```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/type1_data1.csv")
df.head()
```

```python
# 3사분위 1사분위
q1 = df['views'].quantile(0.25)
q3 = df['views'].quantile(0.75)

# IQR 계산
IQR = q3 - q1

# 이상치 기준 찾기
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

# 이상치 조건에 만족하는 데이터 찾기
cond1 = df['views'] < lower_bound
cond2 = df['views'] > upper_bound
df = df[cond1 | cond2]

# 이상치 데이터의 views 컬럼 합
print(int(df['views'].sum()))
```

### 17.이상치, 소수점 있는 데이터 찾기, 표준편차(p168)

### 18.데이터(행) 기준 평균값, 인덱싱(p169)

### 21.시간 간의 차이 계산(분), 필터링(p173)

### 23.시간 간의 차이 계산(분), 비율(p175)

```python
import pandas as pd
# df = pd.read_csv("delivery_time.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/delivery_time.csv")
df.head()
```

### 25.시간 간의 차이 계산(일)(p177)

```python
import pandas as pd
# df = pd.read_csv("delivery_time.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/delivery_time.csv")
df.head()
```

```python
# 각 사용자별로 첫 주문과 마지막 주문 사이의 시간 간격(일 단위)을 계산
df['주문시간'] = pd.to_datetime(df['주문시간'])
min_order_time = df.groupby('user')['주문시간'].min()
max_order_time = df.groupby('user')['주문시간'].max()
time_interval = (max_order_time - min_order_time).dt.days

# 1회 이하 주문 제외
cond1 = time_interval > 0
m = time_interval[cond1].mean()

# 평균보다 기간이 긴 유저 수
cond2 = time_interval > m
print(len(time_interval[cond2]))
```

### 27.시간 범위, 속도(p179)

```python
import pandas as pd
# df = pd.read_csv("delivery_time.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part1/ch3/delivery_time.csv")
df.head()
```

```python
# datetime 자료형으로 변경
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])

# 점심시간 주문 선택
df['시간'] = df['주문시간'].dt.hour
cond1 = df['시간'] >= 10
cond2 = df['시간'] < 13
df = df[cond1 & cond2]

# 속도(km/h) = 거리(km)
df['배달시간'] = df['실제도착시간'] - df['주문시간']
df['배달시간'] = df['배달시간'].dt.total_seconds() / 60 / 60
df['속도'] = df['거리'] / df['배달시간']
# print(df['속도'] >= 50)
print((df['속도'] >= 50).sum())
```
