```python
import pandas as pd
df = pd.read_csv('/content/Titanic.csv')
print(df.shape)
df.head()
```

```python
# Gender, Survived 사이의 **독립성** 검정 -> chi2_contingency
from scipy.stats import chi2_contingency
crosstab = pd.crosstab(df['Gender'], df['Survived'])
crosstab
```

```python
chi2_stats = chi2_contingency(crosstab)
chi2_stats   # 260.71702016732104
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df.columns:
    if df[col].dtype == 'object':
        le = LabelEncoder()
        le.fit(df[col])
        df[col] = le.transform(df[col])
```

```python
# 3-2 target_data -> Survived -> Logit
X = df[['Gender', 'SibSp', 'Parch', 'Fare']]
y = df['Survived']
```

```python
import statsmodels.api as sm
X_const = sm.add_constant(X)
model = sm.Logit(y, X_const).fit()
model.summary()
```

```python
round(model.params['Parch'], 3)
```

```python
# 3-1-2
import numpy as np
odds = np.exp(model.params['SibSp'])
round(odds, 3)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1)
model.fit(tr_x, tr_y)
model.score(tr_x, tr_y), model.score(val_x, val_y)
```

```python
pred = model.predict(test)
result = pd.DataFrame({'pred': pred})
result.to_csv('result.csv', index=False)
```

```python
# 신고접수시간, 출발시간, 도착시간 날짜타입 변환하기
df1 = df.copy()
df1['신고접수시간']= pd.to_datetime(df1['신고접수시간'])
df1['출발시간']= pd.to_datetime(df1['출발시간'])
df1['도착시간']= pd.to_datetime(df1['도착시간'])
print(df1.head())
```
